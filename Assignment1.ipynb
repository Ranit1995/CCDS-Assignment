2) Find commutator relations among Pauli matrix elements. Define all 2x2 Pauli matrices (remember that elements may by imaginary). Define general formalism of commutator calculation using matrix multiplication, and check the commutator relations.

```python
#Defining Pauli matrices

import numpy as np
s1= np.matrix([[0,1],[1,0]])
s2= np.matrix([[0,-1j],[1j,0]])
s3= np.matrix([[1,0],[0,-1]])
print("The Pauli matrices are:")
print(s1)
print(s2)
print(s3)
```

```python
#defining commutation relation

A= np.matrix([[2,7],[2,3]])
B= np.matrix([[3,5],[1,8]])
C= np.dot(A,B)-np.dot(B,A)
print("The commutation relation of A and B is:")
print(C)
```

```python
#Commutation relation of Pauli matrices

C1= np.dot(s1,s2)-np.dot(s2,s1)
C2= np.dot(s2,s3)-np.dot(s3,s2)
C3= np.dot(s3,s1)-np.dot(s1,s3)
print("The commutation relation of s1 and s2 is:")
print(C1)
print("The commutation relation of s2 and s3 is:")
print(C2)
print("The commutation relation of s3 and s1 is:")
print(C3)
```

3) Solve general system of linear equation using matrix inversion approach.

```python
# 3x-5y+2z = 7
# -4x+7y-6z = -21
# 2x+6y-3z = 9

import numpy as np
A= np.array([[3,-5,2],[-4,7,-6],[2,6,-3]])
B= np.array([7,-21,9])
C= np.linalg.inv(A).dot(B)
print("The solutions are:")
print(C)
```

4) Solve ordinary differential equation of 1 dimension using forward Euler method.

```python
# dy/dx = -2y

def f(x,y) : return -2*y
x,y,h = -5.0,1.0,.01
X,Y = [x],[y]

while x < 5.0:
  y = y + h*f(x,y)
  x = x + h

  Y.append(y)
  X.append(x)


import matplotlib.pyplot as plt
plt.plot(X,Y)
plt.show()


  

```
